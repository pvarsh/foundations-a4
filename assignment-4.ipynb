# Literature references
* Suburban spralw cancel carbon savings of cities:
    * News aritcle: http://newscenter.berkeley.edu/2014/01/06/suburban-sprawl-cancels-carbon-footprint-savings-of-dense-urban-cores/
    * Paper (2013): http://pubs.acs.org/doi/abs/10.1021/es4034364 (behind a paywall)
* Large cities are less green (Oliveira et al., Nature):
    http://www.nature.com/srep/2014/140228/srep04235/full/srep04235.html
    * Defines cities using a City Clustering Algorithm (CCA)
    * Finds superlinear size to emissions relationship for CCA cities
    * Includes discussion why MSA are not a good boundary definition:
        * MSAs include large cities only. There are only 274 (388 according to Wikipedia) of them, thus the large population end of the distribution is overrepresented.
        * MSA overestimates the area of small agglomerations
        * ###### Reasoning for why aggregating several small neighboring cities into an MSA may lead to overestimation of efficiency of large cities:
            Imagine several small cities that are each too small to be an MSA and that produce not too much CO2. By aggregating them into an MSA we may end up with an MSA that has a high population and that produces low CO2. The low CO2 production in this case is not due to improved efficiency that is predicted with higher population, but due to efficiency of each of the small city that make up the MSA. In a way the MSA aggregation strips efficient small cities of their accomplishment. Aggregating high CO2 producing cities into an MSA has the opposite effect. To find which of the effects has a higher impact, one has to look at data. This is what this paper claims to have done.
* Growth, innovation, scaling, and the pace of life in cities (Bettencourt et al. et West, PNAS):
    http://www.pnas.org/content/104/17/7301
* A unified theory of urban living (Bettencourt and West, Nature)
    http://www.nature.com/nature/journal/v467/n7318/full/467912a.html
* Are compact cities environmentally friendly?
    http://www.sciencedirect.com/science/article/pii/S0094119012000186
* Does Size Matter? Scaling of CO2 Emissions and U.S. Urban Areas (Fragkias et al. 2013)
    http://www.plosone.org/article/info%3Adoi%2F10.1371%2Fjournal.pone.0064727
    * finds almost exactly the same scaling relation as the calculations below with $\beta_1 = .93$.
    * nice argument against using per capita measurements for power law scaling phenomena.
* Constructing Cities, Deconstructing Scaling Laws (Arcaute, Batty et al. 2014)    
    http://arxiv.org/pdf/1301.1674v2.pdf
    * study of English cities does shows that population is not as strong a predictor as Bettencourt et al. proposed
    * uses clustering algorithm to define city thresholds, discusses MSA and LUZ (Larger Urban Area)
    * (this might not be pertinent) in his talk, West suggested that deviations from his and Bettencourt's theory in England may be explained by economic dominance and centrality of London
* The Area and Population of Cities: New Insights from a Different Perspective on Cities (Rosenfeld et al., American Economic Review, 2011)
    https://www.aeaweb.org/articles.php?doi=10.1257/aer.101.5.2205 (paywall)
    * Clustering. Found that Zipf law holds for cities as small as 5000 in UK, 12000 in US.

###### Population and area data from social explorer

```python
folder = "counties_census_2000_social_explorer"
fname = "R10842464_SL050.csv"
fileIn = dataPath + '/' + folder + '/' + fname

### Read two header lines and build dictionary from first to second
with open(fileIn) as f:
    reader = csv.reader(f)
    line1 = reader.next()
    line2 = reader.next()
    headerDict = {line1[i]: line2[i] for i in range(len(line2))}

    
variablesToKeep = ['FIPS', 'Total Population', 'Urban', 'Population Density (per sq. mile)', 'Area Total', 'Area (Land)']

### read into DataFrame
with open(fileIn, 'r') as f:
    demograph = pd.read_csv(f, header = 1)

### rename columns using first header
demograph.columns = line1

### drop many columns
demograph = demograph[variablesToKeep]

### rename remaining columns
demograph.columns = [u'county_FIPS',
                     u'Total Population',
                     u'Total Population',
                     u'pop_total',
                     u'pop_urban',
                     u'pop_density',
                     u'total_area',
                     u'Area (Land)',
                     u'Area (Land)',
                     u'land_area']

### drop redundant columns
demograph.drop([u'Total Population', u'Area (Land)'], axis = 1, inplace = True)
demograph.head()
```

###### CO2 Emissions data from Vulcan Project

```python
folder = "vulcan_county_CO2_2000"
fname = "CountiesBySector.csv"
fileIn = dataPath + '/' + folder + '/' + fname

with open(fileIn, 'r') as f:
    co2 = pd.read_csv(f)

co2.columns = [u'county_FIPS',
               u'co2_total_mln_t',
               u'commercial',
               u'industrial',
               u'residential',
               u'electricity_prod',
               u'onroad',
               u'cement',
               u'aircraft',
               u'airborne',
               u'nonroad']

### convert co2 output into tonnes
co2['co2_total'] = co2.co2_total_mln_t * 1e6

co2.head()
```

###### County to MSA

```python
folder = "MSA_county_2000"
fname = "tab01.csv"
fileIn = dataPath + '/' + folder + '/' + fname

with open(fileIn, 'r') as f:
    MSA_county = pd.read_csv(f, dtype = str)

### drop columns
MSA_county = MSA_county[['MSA/ CMSA Code', u'FIPS State/ County Code']]
MSA_county.columns = ['msa_code', 'county_FIPS']

# remove empty rows
MSA_county = MSA_county[pd.notnull(MSA_county.iloc[:, 1])]

### convert county_FIPS to np.int64 to match demograph and co2 tables
MSA_county['county_FIPS'] = MSA_county['county_FIPS'].astype(np.int64)

### group by county
grouped = MSA_county.groupby('county_FIPS')
countyToMSA = grouped.first()
countyToMSA['county_FIPS'] = countyToMSA.index

countyToMSA.head()
```

###### MSA to MSA name

```python
folder = "MSA_county_2000"
fname = "tab01.csv"
fileIn = dataPath + '/' + folder + '/' + fname

with open(fileIn, 'r') as f:
    MSA_name = pd.read_csv(f, dtype = str)

### drop columns
MSA_name = MSA_name[['MSA/ CMSA Code', '    Area']]
MSA_name.columns = ['msa_code', 'name']

### only keep MSAs or CMSAs
MSAs = MSA_name['name'].str.extract('([ C]MSA$)')
MSA_name = MSA_name[pd.notnull(MSAs)]

MSA_name.head()
```

```python
### Temporary demographic and co2 join
demo_co2 = pd.merge(demograph, co2, how = 'inner', on = 'county_FIPS')
demo_co2.shape
demo_co2.head()

# Join countyToMSA with MSA_name to lose PMSAs and gain names
# countyToMSA = pd.merge(MSA_name, countyToMSA, how = 'inner', on = 'msa_code')

### Joining temporary table above with county_MSA table and aggregating by MSA
demo_co2_with_MSA = pd.merge(demo_co2, countyToMSA, how = 'inner', on = 'county_FIPS')
grouped = demo_co2_with_MSA.groupby(['msa_code'])
mt_temp = grouped.sum()
mt_temp.drop(['pop_density', 'county_FIPS'], axis = 1, inplace = True) # after summing, these variables are meaningless
mt_temp['msa_code'] = mt_temp.index # for merge

### Join with names
mt = pd.merge(mt_temp, MSA_name, how = 'inner', on ='msa_code')

### Dropping MSAs with total population < 100,000
mt = mt[mt.pop_total > 1e5]
mt['co2_per_capita']=mt.co2_total / mt.pop_total
mt['pop_density'] = mt.pop_total / mt.total_area


#mt = mt.reindex(list(range(mt.shape[0])))

mt.head()
```
