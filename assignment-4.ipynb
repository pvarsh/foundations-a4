# Description of the data and cleaning

## Data files
* **R10842464_SL050.csv**
    * Folder: counties_census_2000_social_explorer
    * From: Social explorer
    * Year 2000
    * Relevant variables
        * Geo_FIPS (county)
        * Total Population
        * Urban Population
        * Area (Total)

* **CountiesBySector.v2.2.web.xls** manually converted to **CountiesBySector.csv**
    * Folder: vulcan_county_CO2_2000
    * From: Vulcan project
    * Year: 2002
    * Relevant variables
        * FIPS (county)
        * Total (CO2 emissions in Million Tonnes)
        * Commercial, Industrial, Residential, etc. emissions
* **tab01_oritinal.xls**
    * Folder: MSA_county_2000
    * From: census.gov
    * Year: 2000
    * Relevant variables
        * MSA/CMSA Code
        * FIPS State/County Code
        * Census population
        * Area *(name)*
        
    Names are processed separately and joined to the table of MSA summaries.

```python
dataPath = "../data"
os.listdir(dataPath)
```

###### Population and area data from social explorer

```python
folder = "counties_census_2000_social_explorer"
fname = "R10842464_SL050.csv"
fileIn = dataPath + '/' + folder + '/' + fname

### Read two header lines and build dictionary from first to second
with open(fileIn) as f:
    reader = csv.reader(f)
    line1 = reader.next()
    line2 = reader.next()
    headerDict = {line1[i]: line2[i] for i in range(len(line2))}

variablesToKeep = ['FIPS', 'Total Population', 'Urban', 'Population Density (per sq. mile)', 'Area Total', 'Area (Land)']

### read into DataFrame
with open(fileIn, 'r') as f:
    demograph = pd.read_csv(f, header = 1)

### rename columns using first header
demograph.columns = line1

### drop many columns
demograph = demograph[variablesToKeep]

### rename remaining columns
demograph.columns = [u'county_FIPS',
                     u'Total Population',
                     u'Total Population',
                     u'pop_total',
                     u'pop_urban',
                     u'pop_density',
                     u'total_area',
                     u'Area (Land)',
                     u'Area (Land)',
                     u'land_area']

### drop redundant columns
demograph.drop([u'Total Population', u'Area (Land)'], axis = 1, inplace = True)
demograph.head()
```

###### CO2 Emissions data from Vulcan Project

```python
folder = "vulcan_county_CO2_2000"
fname = "CountiesBySector.csv"
fileIn = dataPath + '/' + folder + '/' + fname

with open(fileIn, 'r') as f:
    co2 = pd.read_csv(f)

co2.columns = [u'county_FIPS',
               u'co2_total_mln_t',
               u'commercial',
               u'industrial',
               u'residential',
               u'electricity_prod',
               u'onroad',
               u'cement',
               u'aircraft',
               u'airborne',
               u'nonroad']

### convert co2 output into tonnes
co2['co2_total'] = co2.co2_total_mln_t * 1e6
```

###### County to MSA

```python
folder = "MSA_county_2000"
fname = "tab01.csv"
fileIn = dataPath + '/' + folder + '/' + fname

with open(fileIn, 'r') as f:
    MSA_county = pd.read_csv(f, dtype = str)

### drop columns
MSA_county = MSA_county[['MSA/ CMSA Code', u'FIPS State/ County Code']]
MSA_county.columns = ['msa_code', 'county_FIPS']

# remove empty rows
MSA_county = MSA_county[pd.notnull(MSA_county.iloc[:, 1])]

### convert county_FIPS to np.int64 to match demograph and co2 tables
MSA_county['county_FIPS'] = MSA_county['county_FIPS'].astype(np.int64)

### group by county
grouped = MSA_county.groupby('county_FIPS')
countyToMSA = grouped.first()
countyToMSA['county_FIPS'] = countyToMSA.index
```

###### MSA to MSA name

```python
folder = "MSA_county_2000"
fname = "tab01.csv"
fileIn = dataPath + '/' + folder + '/' + fname

with open(fileIn, 'r') as f:
    MSA_name = pd.read_csv(f, dtype = str)

### drop columns
MSA_name = MSA_name[['MSA/ CMSA Code', '    Area']]
MSA_name.columns = ['msa_code', 'name']

### only keep MSAs or CMSAs
MSAs = MSA_name['name'].str.extract('([ C]MSA$)')
MSA_name = MSA_name[pd.notnull(MSAs)]
```

```python
### Temporary demographic and co2 join
demo_co2 = pd.merge(demograph, co2, how = 'inner', on = 'county_FIPS')
demo_co2.shape
demo_co2.head()

# Join countyToMSA with MSA_name to lose PMSAs and gain names
# countyToMSA = pd.merge(MSA_name, countyToMSA, how = 'inner', on = 'msa_code')

### Joining temporary table above with county_MSA table and aggregating by MSA
demo_co2_with_MSA = pd.merge(demo_co2, countyToMSA, how = 'inner', on = 'county_FIPS')
grouped = demo_co2_with_MSA.groupby(['msa_code'])
mt_temp = grouped.sum()
mt_temp.drop(['pop_density', 'county_FIPS'], axis = 1, inplace = True) # after summing, these variables are meaningless
mt_temp['msa_code'] = mt_temp.index # for merge

### Join with names
mt = pd.merge(mt_temp, MSA_name, how = 'inner', on ='msa_code')

### Dropping MSAs with total population < 100,000
mt = mt[mt.pop_total > 1e5]
mt['co2_per_capita']=mt.co2_total / mt.pop_total

mt = mt.reindex(list(range(mt.shape[0])))

mt.head()
```

## Plots

###### Total population against total CO2 emissions

```python
fig1, ax1 = plt.subplots(num = 1, figsize = (10, 6)) 

# Title and axis labels
fig1.suptitle('Total population and CO2 emissions for MSAs', fontsize = 14)
ax1.set_xlabel('log population')
ax1.set_ylabel('log total emissions')

### Scatter
scatter1 = ax1.scatter(np.log(mt.pop_total),
            np.log(mt.co2_total),
            c = "dodgerblue",
            lw = 0,
            s = 30,
            alpha = 0.8,
            label = "MSA")

scatter1_1 = ax1.scatter(np.log(mt.pop_urban),
            np.log(mt.co2_total),
            c = "magenta",
            lw = 0,
            s = 30,
            alpha = 0.4,
            label = "MSA")

### Get x and y lims
xmin, xmax = ax1.get_xlim()
ymin, ymax = ax1.get_ylim()

### Regression
# Design matrix
y = np.log(mt.co2_total)
ones = np.ones(mt.shape[0])
X = np.column_stack((ones, np.log(mt.pop_total)))

# Fit model
mod1 = sm.OLS(y, X).fit()

### Plot regression line
x = np.linspace(xmin, xmax)
y = mod1.params[0] + mod1.params[1] * x
regression_label = r'$\beta_1 = %2.2f$' %mod1.params[1]
line1, = ax1.plot(x, y, c = 'darkorange', lw = 2, label = regression_label)

# Set x and y lims
ax1.set_xlim(xmin, xmax)
ax1.set_ylim(ymin, ymax)

# Legend
extra = ptch.Rectangle((0, 0), 1, 1, fc="w", fill=False, edgecolor='none', linewidth=0)
rsq = r'$R^2 = %2.2f$' %mod1.rsquared
ax1.legend((scatter1, scatter1_1, line1, extra), ("Total population", "Urban population", regression_label, rsq), loc = 2)

# Show
plt.show()
```

###### Total population against per capita CO2 emissions

```python
fig2, ax2 = plt.subplots(num = 2, figsize = (10, 6)) 

# Title and axis labels
fig2.suptitle('Total population and per capita CO2 emissions for MSAs', fontsize = 14)
ax2.set_xlabel('log population')
ax2.set_ylabel('log per capita emissions')

### Scatter
scatter2 = ax2.scatter(np.log(mt.pop_total),
            np.log(mt.co2_per_capita),
            c = "dodgerblue",
            lw = 0,
            s = 30,
            alpha = 0.8,
            label = "MSA")

### Get x and y lims
xmin, xmax = ax2.get_xlim()
ymin, ymax = ax2.get_ylim()

### Regression
# Design matrix
y = np.log(mt.co2_per_capita)
ones = np.ones(mt.shape[0])
X = np.column_stack((ones, np.log(mt.pop_total)))

# Fit model
mod2 = sm.OLS(y, X).fit()

### Plot regression line
x = np.linspace(xmin, xmax)
y = mod2.params[0] + mod2.params[1] * x
regression_label = r'$\beta_1 = %2.2f$' %mod2.params[1]
line2, = ax2.plot(x, y, c = 'darkorange', lw = 2, label = regression_label)

# Set x and y lims
ax2.set_xlim(xmin, xmax)
ax2.set_ylim(ymin, ymax)

# Legend
extra = ptch.Rectangle((0, 0), 1, 1, fc="w", fill=False, edgecolor='none', linewidth=0)
rsq = r'$R^2 = %2.2f$' %mod2.rsquared
ax2.legend((scatter1, line1, extra), ("MSA", regression_label, rsq), loc = 2)

# Show
plt.show()
```
